```python
i = 1

while (i <= 5):
    # calculating space
    j = 5
    while (j > i):
        print("-", end="")
        j -= 1

    # printing stars
    k = 1
    while (k <= i):
        print("*", end=" ")
        k += 1

    print()  # move to next line
    i += 1
```
