#### Assignment
1.write a program to create even number list,odd number list, prime number from the given range.

2.read elements until they exit ex:[1,56,388,19,34.09]


```python
n=int(input("enter your value:"))

even_num=[]
odd_num=[]
prime_num=[]

for num in range(1,n+1,1):
    if (num%2==0):
        even_num.append(num)
    else:
        odd_num.append(num)

    if (num>1): 
        count=0
        for i in range(2,num+1,1):
            if (num%i==0):
                count+=1
        if count == 1:
                prime_num.append(num)

print("Even numbers:", even_num)
print("Odd numbers:", odd_num)
print("Prime numbers:", prime_num)          

```

```python
str1="BeSantTecH1234@78%$#.Com"
up=0
lo=0
vo=0
con=0
num=0
spe=0


for i in str1:
    if i.isupper():
        up+=1
        if i in "AEIOU":
            vo +=1
        else:
            con+=1

    elif i.islower():
        lo+=1
        if i in "aeiou":
            vo+=1
        else:
            con+=1

    elif i.isdigit():
        num+=1


    else:
        spe+=1

print(f"uppercase {up},lowercase {lo},vowels {vo},consonent {con},number {num},special character {spe}")
        

    
```
