##### Slicing in pandas
* iloc(index location) - its based on row/col index value


* var.iloc[row index,col index]
* var.iloc[rst:rsp:rss,cst:csp:css]
* default start value:0
*        stop value: upto last
*        step size:1

* loc - its based on row/col name

```python
# extract 1st rows with 1st 5 columns
mt.iloc[0:5,0:5]
```

```python
# extract entire table with iloc
mt.iloc[:,:]
```

```python
# rows with last 5 columns
mt.iloc[:,-5:]
```

```python
# extract all the col with divisible of 3 row index
mt.iloc[3::3,:]
```

```python
# even index col and odd index rows
mt.iloc[1::2,2::2]
```
